```python
#utils ......................................

def nodeToNums(node):
    if node == None:
        return {}
    if type(node) == set:
        return node
    node = node[1:-1]
    node = node.replace(' ', '')
    nums = node.split(',')
    res = set()
    for n in nums:
        if n.isdecimal:
            res.add(int(n))
    return res


def mergeCycleDicts(c1, c2):
    res = c1.copy()
    for i in c2:
        if not i in c1:
            res[i] = c2[i]
        else:
            for j in c2[i]:
                res[i].append(j)
    return res

def isSubsetOfAny(sets, l):
    for s in sets:
        if l <= s:
            return True
    return False

def makeMergedNodeName(c):
    name = set()
    for n in c:
        name = name | nodeToNums(n)
    name = str(name)
    return name

def getCurrentNode(G, n):
    if n in G:
        return n
    for node in G:
        if nodeToNums(n) < nodeToNums(node):
            return node
    return None
```

```python
# Gets all the cycles of a graph by going into all possible different 
def getCycles(G):
    final_cyc = {}
    visited = set()
    tmp_cyc = {}
    for n in G:
        if not n in visited:
            c = dfsCycle(n, G, [], {}, visited)
            tmp_cyc = mergeCycleDicts(tmp_cyc, c)
    for key in tmp_cyc:
        res = []
        no_flag = False
        for i in tmp_cyc[key]:
            for j in res:
                if set(i) == set(j):
                    no_flag = True
            if not no_flag:
                res.append(i)
        final_cyc[key] = res
    return final_cyc
        

def dfsCycle(n, G, path, cycles, visited):
    visited.add(n)
    path.append(n)
    ipdb.set_trace()
    for nei in G.neighbors(n):
        if nei in path:
            if nei == path[-1]:
                continue
            i = path.index(nei)
            c_size = len(path) - i
            if not (c_size in cycles):
                cycles[c_size] = []
            cycles[c_size].append(path[i:])
            continue
        else:
            dfsCycle(nei, G, path, cycles, visited)
    path.pop()
    return cycles
```

```python
def shrinkCycles(G):
    g2 = G.copy()
    cycles = getCycles(G)
    keys = sorted(cycles.keys(), reverse = True)
    done = []
    for k in keys:
        c_k = cycles[k]
        for c in c_k:
            if isSubsetOfAny(done, c):
                continue
            g2 = shrinkTheCycle(g2, c)
            done.append(c)
    return g2

def shrinkTheCycle(G, c):
    res_G = G.copy()
    merged_node = makeMergedNodeName(c)
    c_in_G_still = False
    res_G.add_node(merged_node)
    res_G.nodes[merged_node]['size'] = len(c)
    for n in c:
        if not n in G:
            continue      
        for ss in G.successors(n):
            if not ss in c:
                ss = getCurrentNode(G, ss)
                res_G.add_edge(merged_node, ss)
        for ps in G.predecessors(n):
            if not ps in c:
                ps = getCurrentNode(G, ps)
                res_G.add_edge(ps, merged_node)
        res_G.remove_node(n)
        c_in_G_still = True
    if not c_in_G_still:
        res_G.remove_node(merged_node)
    return res_G
```
